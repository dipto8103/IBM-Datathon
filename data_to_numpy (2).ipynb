```python
import os
from PIL import Image
import numpy as np
import pickle as pkl

# Path to the main folder containing subfolders of images
main_folder_path = r'D:\Dipto\I.B.M\dermnet(Not Labelled)\train'

# Initialize lists to store image arrays and corresponding labels
image_list = []
labels_list = []

# Walk through the directory tree
for root, dirs, files in os.walk(main_folder_path):
    for filename in files:
        if filename.endswith('.jpg') or filename.endswith('.jpeg'):
            try:
                # Get the full path to the image file
                img_path = os.path.join(root, filename)

                # Extract the folder name (last part of the path) as the label
                label = os.path.basename(root)  # Folder name is the label
                labels_list.append(label)
                
                # Open the image file
                img = Image.open(img_path)

                # Convert image to a NumPy array
                img_array = np.array(img)
                
                # Debug: Print shape of each image and corresponding label
                print(f"Loaded {filename} from {root} with shape {img_array.shape}, label: {label}")
                
                # Append the image array to the list
                image_list.append(img_array)

            except Exception as e:
                # Debug: Print any error encountered
                print(f"Error loading {filename}: {e}")

# Save the image data and labels into .pkl files
with open("labels_list.pkl", 'wb') as labels_file:
    pkl.dump(labels_list, labels_file)

with open("images_list.pkl", 'wb') as images_file:
    pkl.dump(image_list, images_file)
```
